```python
import pycurl
import StringIO

url='https://api.bitfinex.com/v2/candles/trade:1D:tBTCUSD/hist?limit=400'
c=pycurl.Curl()
c.setopt(c.URL, url)

b = StringIO.StringIO()

c.setopt(c.WRITEFUNCTION, b.write)

#c.setopt(c.FOLLOWLOCATION, 1)
#c.setopt(c.HEADER, True) 

c.perform()   
print b.getvalue()   

#b.close()
c.close()
```

```python
import time
```

```python
l1=eval(b.getvalue())
l1
```

```python
for l in l1:
    st = time.localtime(l[0]/1000)
    l[0] = time.strftime('%Y-%m-%d %H:%M:%S', st)       
```

```python
oneday = pd.read_csv("input/one_year.csv")
oneday = oneday[::-1]
cols = ["OPEN", "CLOSE", "HIGH", "LOW"]
oneday[cols] = (oneday[cols] - oneday[cols].mean()) / (oneday[cols].max() - oneday[cols].min())
oneday.shape
```
